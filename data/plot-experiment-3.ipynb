```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import json
```

```python
benchmarks_file = '../implementation/results/experiment-3.json'
with open(benchmarks_file) as d:
    benchmarks = json.load(d)
```

```python
plot_points_linear = []
for benchmark in benchmarks:
    client = benchmark['type']
    chain_size = benchmark['chainSize'] / 365    
    bytesTransferred = benchmark['bytesDownloaded'] + benchmark['bytesUploaded']
    timeToSync = benchmark['timeToSync'] / 1000
    n = benchmark['batchSize' if client != 'superlight' else 'treeDegree']
    plot_points_linear.append([client, chain_size, bytesTransferred, timeToSync, n])
```

```python
points_df = pd.DataFrame(plot_points_linear, columns=['client', 'chain_size', 'bytesTransferred', 'timeToSync', 'n'])
points_df_mean = points_df.groupby(['client', 'chain_size', 'n']).mean().reset_index()
```

```python
points_df_mean
```

```python
op_df = points_df_mean[points_df_mean['client'] == 'optimisticlight']
op_df_20 = op_df[points_df_mean['n'] == 20]
olc_bt_20 = op_df_20['bytesTransferred']
olc_tos_20 = op_df_20['timeToSync']
olc_chain_size_20 = op_df_20['chain_size']

op_df_50 = op_df[points_df_mean['n'] == 50]
olc_bt_50 = op_df_50['bytesTransferred']
olc_tos_50 = op_df_50['timeToSync']
olc_chain_size_50 = op_df_50['chain_size']

op_df_100 = op_df[points_df_mean['n'] == 100]
olc_bt_100 = op_df_100['bytesTransferred']
olc_tos_100 = op_df_100['timeToSync']
olc_chain_size_100 = op_df_100['chain_size']

op_df_200 = op_df[points_df_mean['n'] == 200]
olc_bt_200 = op_df_200['bytesTransferred']
olc_tos_200 = op_df_200['timeToSync']
olc_chain_size_200 = op_df_200['chain_size']

op_df_500 = op_df[points_df_mean['n'] == 500]
olc_bt_500 = op_df_500['bytesTransferred']
olc_tos_500 = op_df_500['timeToSync']
olc_chain_size_500 = op_df_500['chain_size']
```

```python
plt.plot(olc_chain_size_20, olc_tos_20, label ='batch size = 20')
plt.plot(olc_chain_size_50, olc_tos_50, label ='batch size = 50')
plt.plot(olc_chain_size_100, olc_tos_100, label ='batch size = 100')
plt.plot(olc_chain_size_200, olc_tos_200, label ='batch size = 200')
plt.plot(olc_chain_size_500, olc_tos_500, label ='batch size = 500')


plt.xlabel('Chain Size (Years)')
plt.ylabel('Time To Sync (Seconds)')
# plt.xlim((0, 4e6))
plt.xscale('log')
# plt.ylim((0, 60))
plt.legend()
plt.rcParams["figure.figsize"] = plt.rcParamsDefault["figure.figsize"]
plt.show()
```
